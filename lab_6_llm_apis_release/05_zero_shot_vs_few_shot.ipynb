# Zero-Shot vs Few-Shot Prompting

**Zero-shot**: you ask the model to perform a task with **no examples**.

**Few-shot**: you include a **few examples** in the prompt so the model learns the pattern.

| Strategy | Prompt contains | When to use |
|---|---|---|
| Zero-shot | Just the instruction | Simple, well-known tasks |
| One-shot | 1 example | Quick guidance on format/style |
| Few-shot | 2–5 examples | Custom formats, edge cases, domain-specific tasks |

Few-shot prompting is powerful because it teaches the model **by demonstration** rather than by description.

```python
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama',
)

MODEL = "gemma4:e2b"
```

## Task: Generate TF-IDF Quiz Questions

We want the model to generate quiz questions about TF-IDF — with a specific corpus, preprocessing steps, and step-by-step sub-questions.

This is hard to get right with zero-shot because the format is very specific and the model has no idea what we expect.

```python
TASK = "Create 3 TF-IDF quiz questions for an NLP exam."
```

## Zero-Shot

We just describe the task — no examples.

```python
response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": TASK},
    ],
)
print(response.choices[0].message.content)
```

## Few-Shot (One-Shot)

Now we provide **one example** of what a good TF-IDF quiz question looks like. The example is loaded from `assets/5_tf_idf_question.txt` and passed as a user/assistant pair.

```python
with open("assets/5_tf_idf_question.txt") as f:
    example_question = f.read()

response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "user", "content": "Create a TF-IDF quiz question for an NLP exam, using the same structure as the example below.\n\n----\n"   + example_question}
    ],
)
print(response.choices[0].message.content)
```
