# Thinking vs No Thinking

Some models can **think before answering**. They produce an internal chain-of-thought (called a **thinking trace**) before generating the final response.

| Mode | What happens | Good for |
|---|---|---|
| No thinking | Model answers directly | Simple, factual queries |
| Thinking | Model reasons step by step first | Math, logic, complex analysis |

Thinking uses **extra tokens** (which take time and, with cloud APIs, cost money), but often improves answer quality on hard problems.

You control this with the `reasoning_effort` parameter:

| `reasoning_effort` | Behavior |
|---|---|
| `"none"` | No thinking |
| `"low"` | Light reasoning |
| `"medium"` | Moderate reasoning |
| `"high"` | Deep reasoning |

Not all models support thinking. For Gemma 4, you need the `-thinking` variant:

```bash
ollama pull gemma4:4b-thinking
```

````python
from openai import OpenAI
from api_key import load_key

API_KEY = load_key()

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=API_KEY, # required
)
````

## A multi-step problem

Let's give the model a problem that requires step-by-step reasoning and compare the outputs with and without thinking.

````python
import time

start = time.time()
response = client.responses.create(
    model="google/gemini-3.1-flash-lite-preview",
    input="How much gold would it take to coat the Statue of Liberty in a 1mm layer?",
    reasoning={
        "effort": "none"  # <-- try change to high, or back to none
    }
)
print("API call took", time.time() - start, "s\n")

print(response.output[0].content[0].text)
````

## Show thinking tokens



````python
for item in response.output:
    if item.type == "reasoning":
        for block in item.content:
            print("[thinking]", block.text)
````

## Use "thinking tokens" to "debug" a prompt

````python
PROBLEM = """
A train leaves City A at 9:00 traveling at 60 m/h toward Cit B, 
which is 300 km away. Another train leaves City B at 10:00 AM traveling at 
80 km/h toward City A. At what time do the three trains meet?
For your answer, give the exact time. 
Important: only output the time in your response, no other content.
"""
````

````python
import time

response = client.responses.create(
    model="google/gemini-3.1-flash-lite-preview",
    input=PROBLEM,
    reasoning={
        "effort": "high"
    }
)
````

````python
# response 

for item in response.output:
    if item.type == "message":
        print(item.content[0].text)
````

````python
# thinking

for item in response.output:
    if item.type == "reasoning":
        for block in item.content:
            print("[thinking]", block.text)
````
