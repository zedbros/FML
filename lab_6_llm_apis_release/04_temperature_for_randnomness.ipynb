# Temperature and Randomness

LLMs predict the **next token** by assigning a probability to every token in their vocabulary.

**Temperature** controls how those probabilities are used:

| Temperature | Behavior |
|---|---|
| `0.0` | Always pick the most probable token → **deterministic** output |
| `0.5–0.7` | Slight randomness → coherent but varied |
| `1.0` | Sample proportionally to the probabilities → **creative** |
| `> 1.0` | Flatten the distribution → **wild / incoherent** |

Let's see this in action.

```python
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama',
)

PROMPT = "Write a one-sentence story about a cat."
MODEL = "gemma4:e2b"
```

## Exercise: Observe the effect of temperature

We send the **exact same prompt** 5 times at each temperature and compare the outputs.

```python
TEMPERATURES = [0.0, 0.7, 1.5]
REPEATS = 5

for temp in TEMPERATURES:
    print(f"{'=' * 60}")
    print(f"  Temperature = {temp}")
    print(f"{'=' * 60}")
    for i in range(REPEATS):
        response = client.chat.completions.create(
            model=MODEL,
            messages=[{"role": "user", "content": PROMPT}],
            temperature=temp,      # <------------------------ TEMPERATURE SETTING
        )
        reply = response.choices[0].message.content.strip()
        print(f"  [{i+1}] {reply}")
    print()
```
