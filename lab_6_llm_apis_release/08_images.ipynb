# Vision: Sending Images to an LLM

Modern LLMs can process **images** alongside text. This is called **vision** or **multimodal** input.

Instead of sending a simple string as the user message, we send a **list of content parts** — mixing text and images:

```python
messages=[{
    "role": "user",
    "content": [
        {"type": "text", "text": "Describe this image."},
        {"type": "image_url", "image_url": {"url": "data:image/png;base64,..."}}
    ]
}]
```

The image is **base64-encoded** and embedded directly in the API call as a data URL.

````python
import base64
from openai import OpenAI

client = OpenAI(
    base_url='http://localhost:11434/v1',
    api_key='ollama',
)

MODEL = "gemma4:e2b"
````

## Load and encode the image

We read the image file as bytes and convert it to a base64 string, then wrap it in a `data:` URL that the API expects.

````python
IMAGE_PATH = "assets/ironman.jpg"

with open(IMAGE_PATH, "rb") as f:
    image_b64 = base64.b64encode(f.read()).decode()

image_url = f"data:image/png;base64,{image_b64}"
print(f"Image loaded: {len(image_b64)} base64 characters")
````

## Ask the model to describe the image

We send the image alongside a text prompt. The model will analyze the image and return a description.

````python
response = client.chat.completions.create(
    model=MODEL,
    messages=[{
        "role": "user",
        "content": [
            {"type": "text", "text": "Describe what this image shows. Be specific about the structure, labels, and what concept it illustrates."},
            {"type": "image_url", "image_url": {"url": image_url}}
        ]
    }],
)

print(response.choices[0].message.content)
````
