# [Ollama](https://ollama.com/)

Get up and running with large language models locally.

1. Install Ollama on your computer (following the instructions on the website)
2. The models are available at https://ollama.com/search
3. Download the models locally, for example:

```bash
ollama run gemma4:e2b
```

(more info about this model [here](https://ollama.com/library/gemma4))


or if your computer does not have enough memory:

```bash
ollama run gemma3:1b
```


then to be able to use it with the openai lib, start your local API server:

```
ollama serve
```

and then access it like this:

````python
from openai import OpenAI

client = OpenAI(
    base_url = 'http://localhost:11434/v1',
    api_key='ollama', # required, but unused
)

response = client.chat.completions.create(
  model="gemma4:e2b",  # <-- the model from above
  messages=[
    {"role": "system", "content": "Help me create one single quizz question on TF-IDF"},
  ]
)
print(response.choices[0].message.content)
````

(see https://ollama.com/blog/openai-compatibility for more info)
